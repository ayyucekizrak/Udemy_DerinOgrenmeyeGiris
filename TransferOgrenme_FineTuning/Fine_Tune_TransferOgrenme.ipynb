```
from keras.applications.resnet import ResNet50
from keras.preprocessing.image import img_to_array
from keras.applications import imagenet_utils
from PIL import Image
import numpy as np
from io import BytesIO
import os
import requests
```

### ResNet 50 Derin √ñƒürenme Modeli IMAGENET Veri K√ºmesinde Eƒüitildiƒüinde √ñƒürenilen Aƒüƒ±rlƒ±klarƒ±n Transferi ile Test ƒ∞≈ülemi

[ResNet50](https://medium.com/deep-learning-turkiye/deri%CC%87ne-daha-deri%CC%87ne-evri%C5%9Fimli-sinir-a%C4%9Flar%C4%B1-2813a2c8b2a9) Derin √ñƒürenme Modeli ve [ImageNet](https://medium.com/deep-learning-turkiye/motivasyon-yapay-zeka-ve-derin-%C3%B6%C4%9Frenme-48d09355388d) Veri K√ºmesi hakkƒ±nda bilgi i√ßin tƒ±klayƒ±nƒ±z!

```
model = ResNet50(weights="imagenet")
```

```
layers = dict([(layer.name, layer.output) for layer in model.layers])
model.summary()
```

```
# MODELDEKƒ∞ TOPLAM PARAMETRE SAYISINI EKRANA YAZDIR
model.count_params() 
```

```
def prepare_image(image, target):
	# giri≈ü g√∂r√ºnt√ºs√ºn√º yeniden boyutlandƒ±rma ve √∂n i≈ülemerin yapƒ±lmasƒ±
	image = image.resize(target)
	image = img_to_array(image)
	image = np.expand_dims(image, axis=0)
	image = imagenet_utils.preprocess_input(image)

	# i≈ülenmi≈ü g√∂r√ºnt√ºy√º alma
	return image
```

### ƒ∞nternet Kaynaklƒ± bir g√∂r√ºnt√ºn√ºn URL'sini kopyalayarak test i≈ülemini yapabilirsiniz. Binlerce kategorinin olduƒüu ImageNet veri seti sayesinde bir √ßok sƒ±nƒ±fƒ± rahatlƒ±kla kategorize edilebildiƒüini g√∂receksiniz.

```
#@title G√∂r√ºnt√ºn√ºn URL'sini Yapƒ±≈ütƒ±rƒ±n { vertical-output: true }
ImageURL = "https://www.kopekegitimleri.com/wp-content/uploads/2019/07/cavalier-king-charles-spaniel-770x433.jpg\"" #@param {type:"string"}
```

```
#ImageURL = "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQlfqFtPDFn8uLt5A2kR0nBl9NQ9IfaRJmxTA&usqp=CAU"
#https://3.bp.blogspot.com/-u2EcSH2R3aM/VM69jPZvvOI/AAAAAAAAYzk/xmjSdaDD06o/s1600/mercan_resif.jpg
response = requests.get(ImageURL)
image = Image.open(BytesIO(response.content))
image
```

### ƒ∞nternetten Aldƒ±ƒüƒ±nƒ±z Verinin √ñn ƒ∞≈ülemlerinin Yapƒ±lmasƒ± Yeniden Boyutlandƒ±rƒ±lmasƒ± ve Olasƒ± ilk 5 Tahmin ve Tahmin Oranlarƒ±nƒ±n Ekrana Yazdƒ±rƒ±lmasƒ±

√ñrnekteki g√∂rsel i√ßin %91.9 olasƒ±lƒ±kla **mercan** olduƒüunu %0.17 olasƒ±lƒ±kla **deniz≈üakayƒ±ƒüƒ±** ve diƒüer olasƒ±lƒ±klarƒ± h√ºcrenin √ßƒ±ktƒ±sƒ±ndan takip edebilirsiniz.

```
data = {"success": False}

pre_image = prepare_image(image, target=(224, 224)) # 224 x 224 boyutlu hale getir

preds = model.predict(pre_image) # Kesirim modeline √∂n i≈ülemden ge√ßmi≈ü g√∂r√ºnt√ºy√º uygula

results = imagenet_utils.decode_predictions(preds) #kestirim
data["predictions"] = []


for (imagenetID, label, prob) in results[0]: # ImageNet veri k√ºmseinden etiket, olasƒ±lƒ±k ve kestrim sonucunu al
  r = {"label": label, "probability": float(prob)}
  data["predictions"].append(r)
  
data["success"] = True

print(data)
```

### En y√ºksek olasƒ±lƒ±klƒ± sonucun ekrana yazdƒ±rƒ±lmasƒ±

```
print("Sƒ±nƒ±flandƒ±rma tahmini en y√ºksek olan {0} oranƒ±yla {1}'dƒ±r.".format(data["predictions"][0]["probability"],data["predictions"][0]["label"])) 
# En y√ºksek olasƒ±lƒ±klƒ± sonucu ekrana yazdƒ±r
```
